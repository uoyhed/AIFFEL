```python
#손글씨 분류
'''
import pandas as pd

Feature Data 지정하기
digits_df = pd.DataFrame(data=digits.data, columns=digits.feature_names)

Label Data 지정하기
digits_label = digits.target

Target Names 출력해 보기
digits.target_names

데이터 Describe 해 보기
print(digits.DESCR)
'''
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

digits = load_digits()

X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2,random_state=7)

'''
# Decision Tree 사용해 보기
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
'''

'''
# Random Forest 사용해 보기
from sklearn.ensemble import RandomForestClassifier
dtc = RandomForestClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
'''


# SVM 사용해 보기
from sklearn import svm
s = svm.SVC()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)


'''
# SGD Classifier 사용해 보기
from sklearn.linear_model import SGDClassifier
s = SGDClassifier()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)
'''

'''
# Logistic Regression 사용해 보기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
'''

print(classification_report(y_pred,y_test))

from sklearn.metrics import accuracy_score #정확도
ac = accuracy_score(y_pred,y_test)
print(ac)
```

```python
#와인분류
'''
import pandas as pd

Feature Data 지정하기
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)

Label Data 지정하기
wine_label = wine.target

Target Names 출력해 보기
wine.target_names

데이터 Describe 해 보기
print(wine.DESCR)
'''
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

wine = load_wine()

X_train,X_test,y_train,y_test = train_test_split(wine.data,wine.target,test_size=0.2,random_state=7)

'''
# Decision Tree 사용해 보기
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
'''


# Random Forest 사용해 보기
from sklearn.ensemble import RandomForestClassifier
dtc = RandomForestClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)


'''
# SVM 사용해 보기
from sklearn import svm
s = svm.SVC()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)
'''

'''
# SGD Classifier 사용해 보기
from sklearn.linear_model import SGDClassifier
s = SGDClassifier()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)
'''

'''
# Logistic Regression 사용해 보기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
'''

print(classification_report(y_pred,y_test))

from sklearn.metrics import accuracy_score
ac = accuracy_score(y_pred,y_test)
print(ac)
```

```python
#유방암여부
'''
import pandas as pd

# Feature Data 지정하기
bc_df = pd.DataFrame(data=bc.data, columns=bc.feature_names)

# Label Data 지정하기
bc_label = bc.target

# Target Names 출력해 보기
print(bc.target_names)

# 데이터 Describe 해 보기
print(bc.DESCR)
'''
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

bc = load_breast_cancer()

X_train,X_test,y_train,y_test = train_test_split(bc.data,bc.target,test_size=0.2,random_state=7)


# Decision Tree 사용해 보기
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)


'''
# Random Forest 사용해 보기
from sklearn.ensemble import RandomForestClassifier
dtc = RandomForestClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
'''

'''
# SVM 사용해 보기
from sklearn import svm
s = svm.SVC()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)
'''

'''
# SGD Classifier 사용해 보기
from sklearn.linear_model import SGDClassifier
s = SGDClassifier()
s.fit(X_train,y_train)
y_pred = s.predict(X_test)
'''

'''
# Logistic Regression 사용해 보기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
'''

print(classification_report(y_pred,y_test))

from sklearn.metrics import accuracy_score
ac = accuracy_score(y_pred,y_test)
print(ac)
```
